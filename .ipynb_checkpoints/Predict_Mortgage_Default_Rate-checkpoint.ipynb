### 1.4 Encode Categorical Variables to Numeric <a class="anchor" id="1.4"></a>
* In this project, I decide to cut off variables by 10 levels.
* For variables having less than 10 levels, I will use dummy.
* For variables having more than 10 levels, I will use WOE to encode categorical variables.

```python
# List all categorical variables
categorical = ['AP006', 'MB007', 'CD008NEW', 'CD018NEW',
       'CD115NEW', 'CD117NEW', 'CD118NEW', 'CD120NEW', 'CD121NEW', 'CD123NEW',
       'CD130NEW', 'CD131NEW', 'CD132NEW', 'CD133NEW', 'CD135NEW', 'CD136NEW',
       'CD137NEW', 'CD152NEW', 'CD153NEW', 'CD162NEW', 'CD164NEW', 'CD172NEW',
       'CD173NEW', 'CD071NEW', 'CD072NEW', 'CD088NEW', 'CD100NEW', 'CD101NEW',
       'CD106NEW', 'CD107NEW', 'CD108NEW', 'CD113NEW', 'CD114NEW', 'CD160NEW',
       'CD166NEW', 'CD167NEW', 'CD169NEW', 'CD170NEW', 'PA022NEW', 'PA023NEW',
       'PA028NEW', 'PA029NEW', 'PA030NEW', 'PA031NEW']
cate_df = df2.loc[:, categorical]

# Check levels and implement the cut off point: 10 levels
levels = pd.DataFrame(cate_df.nunique()).reset_index()

levels.columns = ['var', 'level']

to_dummy = levels[levels['level']<10]
to_woe = levels[levels['level'] >= 10]

# Get the specific groups of variables
dummy = to_dummy['var'].tolist()
woe = to_woe['var'].tolist()
```

```python
#######################
# Function for WOE #
#######################
def WOE(var):
    k = df2[[var,'loan_default']].groupby(var)['loan_default'].agg(['count','sum']).reset_index()
    k.columns = [var,'Count','Good']
    k['Bad'] = k['Count'] - k['Good']
    k['Good %'] = (k['Good'] / k['Good'].sum()*100).round(2)
    k['Bad %'] = (k['Bad'] / k['Bad'].sum()*100).round(2)
    k[var+'_WOE'] = np.log(k['Good %'] / k['Bad %']).round(2)
    k = k.sort_values(by=var+'_WOE')
    return(k)

k = WOE('CD008NEW')
k.head()
```

```python
# Encode categorical variables respectively
df3 = df2.copy()

for var in woe:
    # Get WOE scores
    k = WOE(var)
    # Append scores back to the dataset
    k1 = k[[var, var+'_WOE']]
    df3 = pd.merge(df3, k1, left_on = var, right_on = var, how = 'left')
    
# Drop the previous columns
df3 = df3.drop(columns = woe)

df3.head()
```

#### 1.5.2 Check null values and replace with mean <a class="anchor" id="1.5.2"></a>

```python
# Check missing values
def null_cell(df):
    missing_values = df.isnull().sum()
    mvp = df.isnull().sum()/df.isnull().count()
    null_values = pd.concat([missing_values, mvp],axis=1,keys=['count','percentage'])
    null_values = null_values.sort_values('count', ascending=False)
    return null_values[null_values['count'] > 0]

fill_list = (null_cell(df3)).index
fill_list
```

```python
# fill the null with mean
for var in fill_list:
    df3.loc[:,var].fillna(df3.loc[:,var].mean(),inplace=True)

df3.describe()
```

### 4.1 Summary Plots <a class="anchor" id="4.1"></a>


```python
def ABS_SHAP(df_shap,df):
    #import matplotlib as plt
    # Make a copy of the input data
    shap_v = pd.DataFrame(df_shap)
    feature_list = df.columns
    shap_v.columns = feature_list
    df_v = df.copy().reset_index().drop('index',axis=1)
    
    # Determine the correlation in order to plot with different colors
    corr_list = list()
    for i in feature_list:
        b = np.corrcoef(shap_v[i],df_v[i])[1][0]
        corr_list.append(b)
    corr_df = pd.concat([pd.Series(feature_list),pd.Series(corr_list)],axis=1).fillna(0)
    # Make a data frame. Column 1 is the feature, and Column 2 is the correlation coefficient
    corr_df.columns  = ['Variable','Corr']
    corr_df['Sign'] = np.where(corr_df['Corr']>0,'red','blue')
    
    # Plot it
    shap_abs = np.abs(shap_v)
    k=pd.DataFrame(shap_abs.mean()).reset_index()
    k.columns = ['Variable','SHAP_abs']
    k2 = k.merge(corr_df,left_on = 'Variable',right_on='Variable',how='inner')
    k2 = k2.sort_values(by='SHAP_abs',ascending = True)
    colorlist = k2['Sign']
    ax = k2.plot.barh(x='Variable',y='SHAP_abs',color = colorlist, figsize=(6,4),legend=False)
    ax.set_xlabel("SHAP Value (Red = Positive Impact)")
    
ABS_SHAP(shap_values,X_test)  
```

* The summary plot shows the variable importance and the SHAP values of every feature for every sample.

* The vertical location shows the feature importance: AP003 is the most important variable, while TD010 is the least important variable.

* The horizonal location shows whether the effect of the value caused a higher or lower prediction, along with the color.
